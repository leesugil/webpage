# k-Means Clustering

**$k$-means clustering** is a method to form $k$ clusters of a given unlabeled numerical data embeded in the feature space, where we assume the choice of $k$ was made a priori.

Here's the algorithm:
- Suppose we have a training set $\{ x^{(1)}, ..., x^{(n)} \}$ and a hypothesis of $k$ number of clusters.
- Initial step: Choose some random vectors (means) for each of the $k$ clusters. (These will converge to what we desire to see.)
- Interation (until convergence):
    - For each $x^{(i)}$, compute the distance to the $k$ mean vectors and choose the closest one.
    - Then the $k$ clusters now have the assigned data. Compute the $k$ mean vectors, each computed from those training data assigned to each cluster.

Let's see this algorithm in action.

We'll generate our data for the presentation purpose. We shall first generate two 100-Gaussian-data groups with means at (1, 1) and (5, 5).

Because we are working on unlabeled data, there's no a priori color code distinguishing the two.

```python
# k-Means clustering
import numpy as np
import matplotlib.pyplot as plt

n = 100
x = np.random.randn(2*n, 2)
x[:n, :] += 1
x[n:, :] += 5
plt.plot(x[:, 0], x[:, 1], 'ko')
plt.show()
```

Assuming that there should be $k=2$ clusters, choose two random initial centers.

I'll pick (2, 3) and (6, -1) as the initial centroids.

```python
mu1 = [2, 3]
mu2 = [6, -1]

plt.plot(x[:, 0], x[:, 1], 'ko')
plt.plot(mu1[0], mu1[1], 'rx', markersize = 12)
plt.plot(mu2[0], mu2[1], 'bx', markersize = 12)
plt.show()
```

Now here comes the iteration part.

We'll iterate five times here. Note that it converges at the 4th iteration.

```python
# Repeat
for i in range(5):
    y = ((x - mu1)**2).sum(1) < ((x - mu2)**2).sum(1) # Is x closer to mu1?
    # Update cluster members
    x1 = x[y > 0.5]
    x2 = x[y < 0.5]
    plt.plot(x1[:, 0], x1[:, 1], 'rs')
    plt.plot(x2[:, 0], x2[:, 1], 'b*')
    # Update new means
    mu1 = x1.mean(0) # (0) means the second index is preserved, thus the mean is
    mu2 = x2.mean(0) # given in the [a, b] form.
    plt.plot(mu1[0], mu1[1], 'kx', markersize = 12)
    plt.plot(mu2[0], mu2[1], 'kx', markersize = 12)
    plt.show()
    print('Iteration ', i+1)
```
